```python
import pandas as pd
le= pd.read_csv("Life Expectancy Data.csv")
lee=pd.DataFrame(le, columns=['Country','Year','LE'])
le1=lee.groupby(['Country'],as_index=False).agg({'LE': 'mean'})
f=pd.read_csv("Fat_Supply_Quantity_Data.csv")
f1=pd.DataFrame(f, columns=['Country','Animal fats','Fruits - Excluding Wine','Obesity'])
inner_join = pd.merge(f1,  
                      le1,  
                      on ='Country',  
                      how ='inner') 
b3=inner_join
b3.head()
```
