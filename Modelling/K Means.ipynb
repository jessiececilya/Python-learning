```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Generate random variables
x = -2 * np.random.rand(200,2)
x0 = 1 + 2 * np.random.rand(100,2)
x[100:200, :] = x0
```

```python
#to observe data points partitioned into two clusters
plt.scatter(x[ : , 0], x[ :, 1], s = 25, color='r')
plt.grid()
```

```python
#Perform K-Means; to establish centroid for cluster
Kmean = KMeans(n_clusters=2)
Kmean.fit(x)
```

```python
#To obtain position of centroid
Kmean.cluster_centers_
```

```python
#Plot the center
plt.scatter(x[:,0], x[:,1], s=25, color='red')
plt.scatter(1.93417391, 1.99971506, s=200, color= 'green')
plt.scatter(-1.04744347, -1.02353328, s=200, color= 'blue')
plt.grid()
plt.show()
```
